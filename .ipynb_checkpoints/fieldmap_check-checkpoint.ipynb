```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define paths
data_dir = '/home/sapje1/data_sapje1/projects/ray/prismafieldmap/21_12_15-12_51_41-STD-1_3_12_2_1107_5_2_43_66075/scans/'
f_phs = 'phase_bet.nii.gz'
f_B = 'fmap1.nii.gz'
```

```python
def load_nifti(file_name):
    import nibabel as nib
    img = nib.load(file_name)
    img_data = img.get_fdata()
    print(img_data.shape)
    return img_data
```

```python
phase_img = load_nifti(data_dir + f_phs)
fmap_img = load_nifti(data_dir + f_B)
```

```python
nib.viewers.OrthoSlicer3D(phase_img).set_position(48,48,32)
nib.viewers.OrthoSlicer3D(fmap_img).set_position(48,48,32)
```

```python
# reshape to 1d for plotting
phase_1d =  np.reshape(phase_img[:,:,:,0], (phase_img[:,:,:,0].size))
print(phase_img[:,:,:,0].shape)
print(phase_1d.shape)
fmap_1d = np.reshape(fmap_img[:,:,:,0], (fmap_img[:,:,:,0].size))
print(fmap_img[:,:,:,0].shape)
print(fmap_1d.shape)
```

```python
# plot of all data shows linear correlation between phase (scanner 
#  image AU) and fsl_prepare_image (rad/s), with discontinuties at 
#  2*pi phase boundaries
rads_Hz = 1.0 / ( 2 * 3.1415927)

fig, ax = plt.subplots()
ax.plot(phase_1d, rads_Hz * fmap_1d, '.')
ax.set_xlabel('phase')
ax.set_ylabel('fmap (Hz)')
```

```python
# mask to include only the unwrapped phase data, arbitrarily
#  +/- 1000 rad/s
fmap_1d_msk = fmap_1d[ (-1000 < fmap_1d) & (fmap_1d < 1000) & (fmap_1d != 0)]
phase_1d_msk = phase_1d[ (-1000 < fmap_1d) & (fmap_1d < 1000)& (fmap_1d != 0) ]
```

```python
# plot of masked data shows linear correlation between phase (scanner 
#  image AU) and fsl_prepare_image (rad/s)
fig, ax = plt.subplots()
ax.plot(phase_1d_msk, fmap_1d_msk, '.')
ax.set_xlabel('phase')
ax.set_ylabel('fmap (rad/s)')
```

```python
# linear fit of the phase data against the fieldmap
#  results in 
np.polyfit(phase_1d_msk, fmap_1d_msk, 1)
#  here slope is ~0.3 (rad/s)/phase
#  total range of the NIFTI is +/-4096 for 2*pi
```
